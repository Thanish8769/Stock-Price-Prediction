```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("D:\Downloads\stockdata_amzn.csv")

df
```

```python
x = df[["High","Low","Open","Volume"]].values

y = df[["Close"]].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()

regressor.fit(x_train, y_train)
```

```python
print(regressor.coef_)

print(regressor.intercept_)

```

```python
predicted = regressor.predict(x_test)

print(predicted)
```

```python
actualvspred = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted' : predicted.flatten()})

actualvspred
```

```python
next_day = x[-1] + 1


next_day = next_day.reshape(1,-1)


next_day_price = regressor.predict(next_day)


print(next_day_price)



```
